```python
player_name=df.loc[df['runs'].idxmax(),'player']
team_name=df.loc[df['runs'].idxmax(),'team']
highest_runs=df.loc[df['runs'].idxmax(),'runs']
opponent=df.loc[df['runs'].idxmax(),'opponent']

print("----> Highest runs given by the bowler '",player_name, "' from the team of '", team_name, "' in one spell which are ",highest_runs, "against team ",opponent)

player_name=df.loc[df['wickets'].idxmax(),'player']
team_name=df.loc[df['wickets'].idxmax(),'team']
wickets=df.loc[df['wickets'].idxmax(),'wickets']
opponent=df.loc[df['wickets'].idxmax(),'opponent']

print("----> Highest wickets taken by the bowler '",player_name, "' from the team of '", team_name, "' in one spell which are ",wickets, "against team ",opponent)

player_name=df.loc[df['6s'].idxmax(),'player']
team_name=df.loc[df['6s'].idxmax(),'team']
highest_6s=df.loc[df['6s'].idxmax(),'6s']
opponent=df.loc[df['6s'].idxmax(),'opponent']
print("----> Highest 6s conceaded by the bowler '",player_name, "' from the team of '", team_name, "' in one spell which are ",highest_6s, "against team ",opponent)

player_name=df.loc[df['4s'].idxmax(),'player']
team_name=df.loc[df['4s'].idxmax(),'team']
highest_4s=df.loc[df['4s'].idxmax(),'4s']
opponent=df.loc[df['4s'].idxmax(),'opponent']
print("----> Highest 4s conceaded by the bowler '",player_name, "' from the team of '", team_name, "' in one spell which are ",highest_4s, "against team ",opponent)

player_name=df.loc[df['maidens'].idxmax(),'player']
team_name=df.loc[df['maidens'].idxmax(),'team']
highest_maidens=df.loc[df['maidens'].idxmax(),'maidens']
opponent=df.loc[df['maidens'].idxmax(),'opponent']
print("----> Highest maidens were delivered by the bowler '",player_name, "' from the team of '", team_name, "' in one spell which are ",highest_maidens, "against team ",opponent)
```

```python
# group the players by the overs, runs, wickets.....
df_player_overs=df.groupby('player').agg({'overs':'sum', 'maidens':'sum', 'runs':'sum', 'wickets':'sum', 'run_rate':'mean', '4s':'sum', '6s':'sum', 'wd':'sum'}).reset_index()
```

```python
# sort the dataset by the overs
sorted_df_player_overs = df_player_overs.sort_values(by='overs', ascending=False).reset_index()
```

```python
sorted_df_player_overs
```

# Plotting the graph between Bowlers & Overs Bowled by them

```python
df_player_overs_top_30 = sorted_df_player_overs.head(30)
# Plotting
plt.figure(figsize=(20, 6))
plt.bar(df_player_overs_top_30['player'], df_player_overs_top_30['overs'])
plt.xlabel('Player')
plt.ylabel('overs')
plt.title('overs by the top 30 players')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()

most_overs_bowled = df_player_overs_top_30.loc[df_player_overs_top_30['overs'].idxmax(), 'player']
most_overs_bowled_value = df_player_overs_top_30['overs'].max()
print("Player With Most Overs Bowled :", most_overs_bowled, "---", most_overs_bowled_value) 
```

# Runs Conceaded by Top 30 Bowlers

```python
df_player_overs_top_30 = sorted_df_player_overs.head(30)

# Plotting
plt.figure(figsize=(15, 6))
plt.bar(df_player_overs_top_30['player'], df_player_overs_top_30['runs'], color='violet')
plt.xlabel('Player')
plt.ylabel('runs')
plt.title('Runs Conceaded by Top 30 Bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()

highest_runs_player = df_player_overs_top_30.loc[df_player_overs_top_30['runs'].idxmax(), 'player']
highest_runs_value = df_player_overs_top_30['runs'].max()

print("Player with the highest runs Conceaded :", highest_runs_player, "-- Runs given:", highest_runs_value)
```

# Run Rate Of The Top 30 Bowlers

```python
df_player_overs_top_30 = sorted_df_player_overs.head(30)

# Plotting
plt.figure(figsize=(15, 6))
plt.bar(df_player_overs_top_30['player'], df_player_overs_top_30['run_rate'], color='orange')
plt.xlabel('Player')
plt.ylabel('run_rate')
plt.title('Run Rate Of The Top 30 Bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()

highest_run_rate = df_player_overs_top_30.loc[df_player_overs_top_30['run_rate'].idxmax(), 'player']
highest_run_rate_value = df_player_overs_top_30['run_rate'].max()
print("Player With The Highest Run Rate :", highest_run_rate, "---", highest_run_rate_value)
```

# Most Wide Balls Delivered By Top 30 Bowlers

```python
df_player_overs_top_30 = sorted_df_player_overs.head(30)

# Plotting
plt.figure(figsize=(15, 6))
plt.bar(df_player_overs_top_30['player'], df_player_overs_top_30['wd'], color='pink')
plt.xlabel('Player')
plt.ylabel('Wide balls')
plt.title('Most Wide Balls Delivered By Top 30 Bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()

most_wide_balls = df_player_overs_top_30.loc[df_player_overs_top_30['wd'].idxmax(), 'player']
most_wide_balls_value = df_player_overs_top_30['wd'].max()
print("Player With The Most Wide Balls :", most_wide_balls, "---", most_wide_balls_value)
```

# Maiden Overs and Overs Delivered By Top 30 Bowlers

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot bars for 'overs' on the primary y-axis (ax1)

index = np.arange(len(df_player_overs_top_30))
bars = ax1.bar(index, df_player_overs_top_30['overs'], label='Overs', color='skyblue')

# Plot line for 'wickets' on the same y-axis (ax1)
line = ax1.plot(index, df_player_overs_top_30['maidens'], color='orange', marker='o', label='maidens')

# Set labels and titles
ax1.set_xlabel('Player')
ax1.set_ylabel('Overs', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')  # Make sure y-axis labels are in blue

plt.title('Maiden Overs and Overs Delivered By Top 30 Bowlers')

# Combine legends
bars_legend = [bar for bar in bars]
line_legend = [line[0]]
plt.legend(bars_legend + line_legend, ['Overs', 'Maiden '], loc='upper left')

# Set x-axis ticks and labels
plt.xticks(index, df_player_overs_top_30['player'], rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()

most_maidens = df_player_overs_top_30.loc[df_player_overs_top_30['maidens'].idxmax(), 'player']
most_maidens_value = df_player_overs_top_30['maidens'].max()
print("Player With The Most Maidens :", most_maidens, "---", most_maidens_value)
```
